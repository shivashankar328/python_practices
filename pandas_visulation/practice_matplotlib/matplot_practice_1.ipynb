```python
# making sign curves using np.pi
x3 = np.arange(0, 4 * np.pi, 0.1)
y3 = np.sin(x3)
plt.plot(x3, y3, label='Sin(x)')
plt.title('sign curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
```

```python
# subplots for sin and cos curves
x3 = np.arange(0, 6*np.pi, 0.2)
y_sin = np.sin(x3)
y_cos = np.cos(x3)
plt.subplot(2, 1, 1)
plt.plot(x3, y_sin, 'b--', label="sin x")
plt.title('sin x')
plt.legend()
# cos plot
plt.subplot(2, 1, 2)
plt.plot(x3, y_cos, 'g--', label='cos x')
plt.title('cos x')
plt.legend()
```

```python
# bar plot
x = [2,8,10]
y= [11, 16, 9]

x2 = [3,9, 13]
y2 = [4,8, 11]

plt.bar(x, y)
plt.bar(x2, y2, color='green')
plt.title('bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# histogram 
a = np.array([13, 22, 16, 28, 19, 33, 54, 59, 65, 71, 44, 39, 10])
plt.hist(a, bins=15)
plt.title('histogram')
```

```python
# box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangel box plot
plt.boxplot(data, vert = True, patch_artist=True);

                                                      
```

```python
# box plt horizontal
plt.boxplot(data, vert = False, patch_artist=False);
```
